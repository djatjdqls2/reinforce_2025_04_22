```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 로그 불러오기
a_logs = pd.read_csv('./logs/rewards_q_gamma01.csv', index_col=False).iloc[:, 1]
b_logs = pd.read_csv('./logs/rewards_q_gamma05.csv', index_col=False).iloc[:, 1]
c_logs = pd.read_csv('./logs/rewards_q_gamma09.csv', index_col=False).iloc[:, 1]
d_logs = pd.read_csv('./logs/rewards_q_gamma99.csv', index_col=False).iloc[:, 1]

# 에피소드 인덱스
episodes = range(len(a_logs))

# 그래프 출력
plt.figure(figsize=(16, 8))
plt.plot(a_logs.cumsum() / (pd.Series(np.arange(a_logs.shape[0])) + 1), label="gamma=0.1")
plt.plot(b_logs.cumsum() / (pd.Series(np.arange(b_logs.shape[0])) + 1), label="gamma=0.5")
plt.plot(c_logs.cumsum() / (pd.Series(np.arange(c_logs.shape[0])) + 1), label="gamma=0.9")
plt.plot(d_logs.cumsum() / (pd.Series(np.arange(d_logs.shape[0])) + 1), label="gamma=0.99")
plt.xlabel("Episode")
plt.ylabel("Average Reward")
plt.title("Gamma Comparison")
plt.legend()
plt.grid(True)
plt.savefig("gamma_compare.jpg", dpi=300, bbox_inches='tight')
plt.show()
```
