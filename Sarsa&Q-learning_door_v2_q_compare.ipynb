```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#  
sarsa_logs = pd.read_csv('./logs/rewards_sarsa_door_v2.csv', index_col=False).iloc[:, 1]
q_logs = pd.read_csv('./logs/rewards_qlearning_door_v2.csv', index_col=False).iloc[:, 1]

#  
episodes = range(len(sarsa_logs))

#  
plt.figure(figsize=(16, 8))
plt.plot(q_logs.cumsum() / (pd.Series(np.arange(q_logs.shape[0])) + 1), label="Q-Learning")
plt.plot(sarsa_logs.cumsum() / (pd.Series(np.arange(sarsa_logs.shape[0])) + 1), label="SARSA")
plt.xlabel("Episode")
plt.ylabel("Average Reward")
plt.title("Q-value Comparison")
plt.legend()
plt.grid(True)
plt.savefig("compare_door_v2.jpg", dpi=300, bbox_inches='tight')
plt.show()
```
