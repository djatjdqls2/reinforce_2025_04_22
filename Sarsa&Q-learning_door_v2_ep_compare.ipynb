```python
import pandas as pd
import matplotlib.pyplot as plt

#  CSV  
sarsa_df = pd.read_csv("./logs/rewards_sarsa_door_v2.csv")
qlearn_df = pd.read_csv("./logs/rewards_qlearning_door_v2.csv")

#   
sarsa_rewards = sarsa_df.iloc[:, 1].values
qlearn_rewards = qlearn_df.iloc[:, 1].values

#   ( > 0   )
sarsa_success = (sarsa_rewards > 0).mean() * 100
qlearn_success = (qlearn_rewards > 0).mean() * 100

# 10  
sarsa_sampled = sarsa_rewards[::10]
qlearn_sampled = qlearn_rewards[::10]
x = list(range(len(sarsa_sampled)))  # x   ( )

# 
plt.figure(figsize=(10, 6))
plt.plot(x, sarsa_sampled, label=f"SARSA (Success = {sarsa_success:.1f}%)")
plt.plot(x, qlearn_sampled, label=f"Q-Learning (Success = {qlearn_success:.1f}%)")
plt.title("Episode Rewards (Every 10 Episodes): SARSA vs Q-Learning")
plt.xlabel("Episode (every 10th)")
plt.ylabel("Reward")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("compare_door_ep.jpg", dpi=300, bbox_inches='tight')
plt.show()
```
